```python
import pandas as pd 
import numpy as np 
from datetime import datetime
import pandas_datareader.data as web 
import matplotlib.pyplot as plt 
```

```python
start = datetime(2015,1,1)
end = datetime(2022,1,8)
```

```python
apple = web.DataReader('AAPL', 'yahoo', start, end)
bmw = web.DataReader('BMW.DE', 'yahoo', start, end)
amazon = web.DataReader('AMZN', 'yahoo', start, end)
```

```python
apple.info()
```

```python
apple.describe()
```

```python
plt.style.use('seaborn')
```

```python
apple['Close'].plot(label = 'Apple close', figsize=(16, 9), fontsize =12)
bmw['Close'].plot(label = 'BMW close')
amazon['Close'].plot(label = 'Amazon close')
plt.legend()
plt.title('Close price')
plt.ylabel('points')
#fontsize - размер шрифта
```

```python
amazon['Total Traded'] = amazon['Open']*amazon['Volume']
bmw["Total Traded"] = bmw['Open']*bmw['Volume']
apple["Total Traded"] = apple['Open']*apple['Volume']
apple
```

```python
amazon['Total Traded'].plot(label = 'AMZN Total Traded', figsize =(16,5))
bmw['Total Traded'].plot(label = 'BMW Total Traded')
apple['Total Traded'].plot(label = 'AAPL Total Traded')
plt.legend()
plt.ylabel('Total Traded')
plt.title('Total traded companies')
plt.show()
```

```python
#нахождение максимального элемента из матрицы
apple['Total Traded'].argmax()
```

```python
apple['Close'].plot(figsize=(16,5))
apple['SMA60'] = apple['Close'].rolling(60).mean()
apple['SMA300']= apple['Close'].rolling(300).mean()
plt.ylabel('Points')
plt.title('Moving average')
apple['SMA300'].plot(label = 'SMA300')
apple['SMA60'].plot(label = 'SMA60')
plt.legend()
```

```python
bmw['returns'] = (bmw['Close']/bmw['Close'].shift(1)) - 1
bmw
```

```python
apple['returns'] = (apple['Close']/apple['Close'].shift(1)) - 1
bmw
```

```python
amazon['returns'] = (amazon['Close']/amazon['Close'].shift(1)) - 1
bmw
```

```python
bmw['returns'].hist(bins = 70, label = 'BMW')
```

```python
amazon['returns'].hist(bins = 70)
```

```python
apple['returns'].hist(bins = 70)
```

```python
bmw['returns'].plot(kind = 'kde', label = 'BMW', figsize = (15, 7))
amazon['returns'].plot(kind = 'kde', label = 'Amazon')
apple['returns'].plot(kind = 'kde', label = 'Apple')
plt.legend()
```
