```python
aapl= close.AAPL.copy().to_frame()
aapl
#all apple close from the beginning till the end 
```

```python
aapl.shift(periods = 1)
#передвигает переменную
```

```python
aapl['lag1']= aapl.shift(periods=1)
aapl
#lag - shifts a column down by a certain number - first NaN, after digits
#lead- shifts a column up by a certain number- first digits, after NaN
```

```python
aapl['Diff'] = aapl.AAPL.sub(aapl.lag1)
aapl
#shows the difference between aapl and lag1 
```

```python
aapl['% Change'] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100) #as we have to calculate percentage change 
aapl['Diff2']= aapl.AAPL.diff(periods = 1) #вычитание разницы, how which period you want to different
aapl['% Change 2'] = aapl.AAPL.pct_change(periods=1).mul(100)# 2 ways of calculating percentage change
aapl
#we've got the percentage change on a daily basis 
```

```python
aapl.rename(columns = {'% Change 2': 'Change'}, inplace = True)
aapl
```

```python
aapl.AAPL.resample('M').last() #gives you last day of that specific month 
```

```python
aapl.AAPL.resample('BM').last().pct_change(periods =1).mul(100) # we got the centre change on a monthly frame 
```

```python
del aapl['Diff2']
del aapl['Diff']
del aapl['lag1']
del aapl['Change']
del aapl['% Change']
aapl
```

```python
ticker = ['AAPL', 'SPY', 'KO', 'IBM', 'DIS', 'MSFT']
stocks = yf.download(ticker, start = '2010-01-01', end = '2022-01-01')
```

```python
close = stocks.loc[:, 'Close'].copy()
```

```python
normclose= close.div(close.iloc[0]).mul(100)
normclose.plot(figsize = (15,8), fontsize=12 )
plt.legend(fontsize = 12)
plt.show()
```

```python
ret = close.pct_change().dropna()
ret.head()
```

```python
ret.describe()
```

```python
ret.describe().T #T= transpose, change axes 
```

```python
summary = ret.describe().T.loc[:, ['mean', 'std']]
summary 
```

```python
summary['mean']= summary['mean']*252
summary['std'] = summary['std']*np.sqrt(252)
summary
```

```python
#std - standard deviation is the statistical measure of market volatility, measuring how widely prices are disperced from the average price
#if prices trade in a narrow trading range, std will return a low value that indicates low volatility
#spy=sp500
summary.plot.scatter(x = 'std', y='mean', figsize=(12,8), s = 50, fontsize = 15)
for i in summary.index:
    plt.annotate(i, xy = (summary.loc[i, 'std']+ 0.002, summary.loc[i, 'mean']+ 0.002), size = 15)
plt.xlabel('Annual risk(std)', fontsize = 15) 
plt.ylabel('Annual return', fontsize = 15)
plt.title('Risk/return', fontsize = 25)
```
